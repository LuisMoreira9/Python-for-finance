# <center>ARRAYS OF DATA</center>

Lists are very flexible, but can also use a lot of memory, and even have alower performance

To fix this we have the array - arrays structure other objects of the same data type in rows and columns

If we think only of numbers, 

A 1-dimensional array is a vector of real numbers (single row or column of elements only)

In more common cases, the 2-dimensional array represents an i x j matrix of elements

This concept generalizes to i x j x k cubes of elements in 3 dimensions

As well as to general n-dimensional arrays of shape i x j x k x l x ...

It is useful to have a specialized class of data structures explicitly desgined to handle arrays conveniently and efficiently. 

This is where the NumPy library comes into play, with its powerful ndarray class

# <center>THE PYTHON ARRAY CLASS </center>

There is a dedicated array module in python

This module defines an object type which can compactly represent an array of basic
values: characters, integers, floating point numbers. Arrays are sequence types and
behave very much like lists, except that the type of objects stored in them is con‐
strained. The type is specified at object creation time by using a type code, which is a
single character.

```python
v = [0.5, 0.75, 1.0, 1.5, 2.0]

import array

a=array.array('f', v)
a
```

```python
a.append(0.5)
a
```

```python
a.extend([5.0, 6.75])
a
```

```python
2*a
```

In this case only float objects can be appended. Other types may raise errors

Another useful advantage of arrays is:

```python
with open('array.apy', 'wb') as f:
    a.tofile(f)
```

There are several ways to initialize a ndarray object.
One is via np.array, however this assumes all elements of the array are already available.

In contrast, one might like to have the ndarray objects initialized first (empty) to populate them later with results generated during the execution of code.
For this, we can use the following functions

```python
c=np.zeros((2,3), dtype='i', order='C')
c
```

```python
c = np.ones((2, 3, 4), dtype='i', order='C') 
c
```

```python
d = np.zeros_like(c, dtype='float16', order='C') #differs from np.zeros because it takes the shape of c instead of the shape of the arguments
d
```

```python
e = np.empty((2, 3, 2)) 
e
```

```python
f = np.empty_like(c) 
f
```

```python
np.eye(5)
```

```python
np.linspace(5, 15, 12)
```

SHAPING AND RESIZING

Although ndarray objects are immutable by default, there are multiple options to reshape and resize them.
While reshaping in general just provides another view on the same data, resizing in general creates a new (temporary) object.

```python
g=np.arange(15)
g
```

```python
g.shape
```

```python
np.shape(g)
```

```python
g.reshape((3,5))
```

```python
h=g.reshape((5,3))
h
```

```python
h.T
```

```python
h.transpose()
```

During reshaping, the total number of observations is unchanged. During resizing, it changes:
Increases -- up-sizing
Decreases -- down-sizing

```python
g
```

```python
np.resize(g, (3,1))
```

```python
np.resize(g, (1,5))
```

```python
np.resize(g, (2, 5)) 
```

```python
n = np.resize(g, (5, 4)) 
n
```

STACKING

Operation that allows the horizontal or vertical combination of 2 ndarray objects.
However, the size of the "connecting" dimension must be the same

```python
h
```

```python
np.hstack((h,2*h))
```

```python
np.vstack((h,2*h))
```

FLATTENING

Flattens a multi-dimensional array into a one-dimensional one.
One can choose whether the flattening happens row-by-row (C order) or column-by-column (F order)

```python
h
```

```python
h.flatten()
```

```python
h.flatten(order='C') #this is the default, it flattens row-wise
```

```python
h.flatten(order='F')
```
