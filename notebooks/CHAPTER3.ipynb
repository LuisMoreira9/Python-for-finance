# String Methods in Python

| Method | Arguments | Returns/Result |
|--------|-----------|----------------|
| `capitalize` | `()` | Copy of the string with first letter capitalized |
| `count` | `(sub, start, end)` | Count number of occurrences of substring |
| `encode` | `(encoding, errors)` | Encoded version of the string |
| `find` | `(sub, start, end)` | Lowest index where substring is found |
| `join` | `(seq)` | Concatenation of strings in sequence `seq` |
| `lower` | `()` | Copy with all letters in lowercase |
| `replace` | `(old, new, count)` | Replaces `old` with `new` the first `count` times |
| `split` | `(sep, maxsplit)` | List of words in string with `sep` as separator |
| `splitlines` | `(keepends)` | Separated lines with line ends if `keepends` is True |
| `strip` | `(chars)` | Copy with leading/trailing characters in `chars` removed |
| `upper` | `()` | Copy with all letters in uppercase |

REGULAR EXPRESSIONS

if something is regularly repeated in a string, we can use re to find all instances of it

```python
import re

series = """
 '01/18/2014 13:00:00', 100, '1st';
 '01/18/2014 13:30:00', 110, '2nd';
 '01/18/2014 14:00:00', 120, '3rd'
 """

dt= re.compile("'[0-9/:\s]+'") #datetime
result=dt.findall(series)
print(result)

#now we can parse the str objects containing the datetime info 

from datetime import datetime
pydt=datetime.strptime(result[0].replace("'",""), '%m/%d/%Y %H:%M:%S')
print(pydt)

a='aaaaaaa; bbbbbbbb; aaaaaaa; cccccccc; aaaaaaa'
onlya= re.compile('a{7}') #only 7 a's
resultsofa=onlya.findall(a)
print(resultsofa)
```
