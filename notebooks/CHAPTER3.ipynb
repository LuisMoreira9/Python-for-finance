REGULAR EXPRESSIONS

if something is regularly repeated in a string, we can use re to find all instances of it

```python
import re

series = """
 '01/18/2014 13:00:00', 100, '1st';
 '01/18/2014 13:30:00', 110, '2nd';
 '01/18/2014 14:00:00', 120, '3rd'
 """

dt= re.compile("'[0-9/:\s]+'") #datetime
result=dt.findall(series)
print(result)

#now we can parse the str objects containing the datetime info 

from datetime import datetime
pydt=datetime.strptime(result[0].replace("'",""), '%m/%d/%Y %H:%M:%S')
print(pydt)

a='aaaaaaa; bbbbbbbb; aaaaaaa; cccccccc; aaaaaaa'
onlya= re.compile('a{7}') #only 7 a's
resultsofa=onlya.findall(a)
print(resultsofa)
```

| Method          | Arguments               | Returns/Result                                          |
|-----------------|-------------------------|---------------------------------------------------------|
| `l[i] = x`      | `[i]`                   | Replaces i-th element by x                             |
| `l[i:j:k] = s`  | `[i:j:k]`               | Replaces every k-th element from i to j-1 by s         |
| `append`        | `(x)`                   | Appends x to object                                     |
| `count`         | `(x)`                   | Number of occurrences of object x                       |
| `del l[i:j:k]`  | `[i:j:k]`               | Deletes elements with index values i to j-1 and step k  |
| `extend`        | `(s)`                   | Appends all elements of s to object                     |
| `index`         | `(x[, i[, j]])`         | First index of x between elements i and j-1             |
| `insert`        | `(i, x)`                | Inserts x at/before index i                             |
| `remove`        | `(x)`                   | Removes element x at first match                        |
| `pop`           | `(i)`                   | Removes element with index i and returns it             |
| `reverse`       | `()`                    | Reverses all items in place                             |
| `sort`          | `([cmp[, key[, reverse]]])` | Sorts all items in place                            |

DIFFERENCES BETWEEN TUPLE AND LIST

| Feature    | Tuple                              | List                        |
|------------|------------------------------------|-----------------------------|
| Syntax     | `t = (1, 2.5, 'data')`             | `l = [1, 2.5, 'data')`      |
| Mutable?   | ❌ No (can't change after creation) | ✅ Yes (can add/remove/edit) |
| Speed      | Faster                             | Slower                      |
| Use case   | Fixed data                         | Data that changes           |
