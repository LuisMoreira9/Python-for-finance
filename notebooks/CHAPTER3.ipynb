# <center>REGULAR EXPRESSIONS</center>

if something is regularly repeated in a string, we can use re to find all instances of it

```python
import re

series = """
 '01/18/2014 13:00:00', 100, '1st';
 '01/18/2014 13:30:00', 110, '2nd';
 '01/18/2014 14:00:00', 120, '3rd'
 """

dt= re.compile("'[0-9/:\s]+'") #datetime
result=dt.findall(series)
print(result)

#now we can parse the str objects containing the datetime info 

from datetime import datetime
pydt=datetime.strptime(result[0].replace("'",""), '%m/%d/%Y %H:%M:%S')
print(pydt)

a='aaaaaaa; bbbbbbbb; aaaaaaa; cccccccc; aaaaaaa'
onlya= re.compile('a{7}') #only 7 a's
resultsofa=onlya.findall(a)
print(resultsofa)
```

# <center>DIFFERENCES BETWEEN TUPLE AND LIST</center>

| Feature    | Tuple                              | List                        |
|------------|------------------------------------|-----------------------------|
| Syntax     | `t = (1, 2.5, 'data')`             | `l = [1, 2.5, 'data')`      |
| Mutable?   | ❌ No (can't change after creation) | ✅ Yes (can add/remove/edit) |
| Speed      | Faster                             | Slower                      |
| Use case   | Fixed data                         | Data that changes           |

# <center>SETS</center>

Are unordered collections of other objects, containing every element only once

```python
s=set(['u', 'd', 'u', 'd', 'du', 'ud', 'du'])
s
```

```python
l=['u', 'd', 'u', 'd', 'du', 'ud', 'du']
toset=set(l)
toset
```

Turning lists into sets is a way to see all unique values in the list

```python
s=set(['u', 'd', 'u', 'd', 'du', 'ud', 'du'])

t = set(['d', 'dd', 'uu', 'u'])

s.union(t) #items in either s or t but not in both
```

```python
s.intersection(t) #items in both s and t
```

```python
s.difference(t) #items in s but not in t
```

```python
t.difference(s) #items in t but not in s
```

```python
s.symmetric_difference(t) #items in either s or t but not in both
```
